Laplacian Matrix:
Graph: n nodes and represented by an n Ã— n adjacency matrix

- The adjacency matrix for directed graphs is not necessarily symmetric, whereas the adjacency matrix for undirected graphs is symmetric.

```python
from random import sample
import numpy as np

n = 10
nodes = list(np.arange(n))


V = np.zeros(n, dtype = tuple)
for i in range(n):
    nv = np.random.randint(5)

    itself = True
    while itself:
        cnodes = sample(nodes, nv)
        itself = i in cnodes
    V[i] = cnodes

# Test matrix
A = np.zeros((n,n))
for i in range(n):
    for j in range(n):
        if j in V[i]:
            A[i,j] = np.random.random()*5



# Laplacian of network
J = np.copy(A)
for i in range(n):
    J[i,i] = - np.sum(A[:,i]) - A[i,i]
```
