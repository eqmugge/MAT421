## Theory

###Root Finding
Root finding is the mathematical method where an algorithm is used to find zeros or ‘roots’ of continuous functions. Using Python libraries such as scipy.optimize allows solving these roots to be simple and efficient. When considering the potential uses of root finding algorithms in research and prediction of breast cancer, specific characteristics of a mass can be determined such as size, shape, and texture. And by using root finding methods an exact point can be found where the probability of a given mass being cancerous is higher based on the critical characteristics.

#### Root Finding Methods Formulas

**Newton-Raphson Method:**
$$ x_{\text{new}} = x_{\text{old}} - \frac{f(x_{\text{old}})}{f'(x_{\text{old}})} $$

Newton-Raphson Method could potentially be used to find optimal parameters to improve a complex predictive model of breast cancer by minimizing error. This method is most appropriate for differentiable and smooth models as it requires a derivative.

This method can be impracticable if the function in question is overly complex or the derivative is not readily available.

**Bisection Method:**
$$ m = \frac{a + b}{2} $$

Bisection Method Formula is another root finding method to be considered. The bisection does not require a function to be differentiable making it more suitable for use in narrowing down markers such as tumor size to a specific cut off point.

The convergence for Bisection is generally slower than it is for Newton-Raphson it can also become unpredictable if the function is not sufficiently smooth.

**Secant Method:**
$$ x_{\text{new}} = x_1 - \frac{f(x_1) \times (x_1 - x_0)}{f(x_1) - f(x_0)} $$

Secant Method Formula the final root finding method to be considered, the secant method does not require finding difficult derivatives as it uses the values of the function at two points to approximate the derivative.

Secant method has the slowest convergence rate of the three methods as well as requiring the initial interval to be chosen in such a way that the function changes signs over the interval which ultimately makes this the least reliable method.



#### Effectiveness

Although root finding is a powerful numerical method, it is limited by several factors. First, the underlying function of the data has to be "smooth" or continuous. Second, convergance is not guaranteed for overly complex data sets. Due to the nature of breast cancer data, which is complex and also irregular we cannot use this method as we had hoped in creating a predictive model.

###Numerical Differentiation
Numerical differentiation is a computational method used to approximate the derivatives of a function from its discrete values. This method can be incredibly effective in breast cancer research, where it aids in the analysis of time-series or sequential data points that represent fluctuations in biological markers, tumor dimensions, or other measurable clinical variables over time. By assessing the rate of change of these critical features, numerical differentiation provides insight into the dynamics of tumor growth and disease progression, offering a perspective that static data alone cannot provide.
### Numerical Differentiation formulas

**Forward Difference Formula:**

$$f'(x_j) \approx \frac{f(x_{j+1}) - f(x_j)}{h}$$
The forward difference formula is apllicable for sequential data points that represent changes over time. It could be used in breast cancer research to analyze the rate of change in biological markers over successive time intervals.

**Backwards Difference Formula:**

$$f'(x_j) \approx \frac{f(x_j) - f(x_{j-1})}{h}$$

The backward difference formula is applicable for historical trends or retrspective data. It could potentially be used in breast cancer research to assess the effectivness of past treatments or identify patterns in recurrence.

**Central Difference Formula:**

$$f'(x_j) \approx \frac{f(x_{j+1}) - f(x_{j-1})}{2h}$$
The central difference formula is applicable for both forward and backward data points by considering both in the formula. This formula can be used to capture changes with greater accuracy than the other two and can be used to analyze more subtle changes in biomarker levels or tumor dimensions.

Utilizing libraries like NumPy for basic array operations and SciPy for more intricate functions, including direct derivative calculations, streamlines the process, enhancing efficiency and accuracy. This capability is crucial for effectively processing and analyzing medical data.

Once these differentiated features are extracted, they can be integrated into a variety of machine learning models designed to handle sequential data. Techniques such as time-series analysis, Long Short-Term Memory networks, logistic regression, or Support Vector Machines can leverage this dynamic information. These models are adept at predicting crucial outcomes such as disease progression, patient prognosis, or treatment efficacy, thus significantly improving the impact of medical interventions.

###Effectiveness
Unfortunately, the applicability of numerical differentiation formulas is extremely limited in scenarios where the data is not smooth. Thus in the context of breast cancer data where the dataset exhibits irregularities and outliers distorts the rate of change calculations derived from numerical differentiation. Ultimately, this leads to numerical differentiation being inefficient compared to other mathematical methods such as logistic regression in predicting breast cancer.



###Logistic Regression

Logistic regression is a valuable tool for modeling binary outcomes, especially useful when the dependent variable is categorical. This makes it an ideal mathematical method for estimating the probability that a given observation falls into a specific category.


**Mathematical Model**

Using the logistic function, logistic regression transforms its output into a probability score between 0 and 1, ensuring that the predictions are bounded within this range.

The logistic function is defined as:

$$ p = \frac{1}{1 + e^{-(\beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_n x_n)}} $$

Where:
- $p$ is the probability of the dependent event occurring.
- $\beta_0, \beta_1, \ldots, \beta_n$ are the coefficients of the model.
- $x_1, x_2, \ldots, x_n$ are the independent variables.


**Application**

Logistic regression is commonly utilized in the medical field due to its effectiveness in handling binary classification problems. It is frequently employed in diagnostic tests which yield binary outcomes—positive or negative. For instance, in breast cancer prediction, logistic regression can analyze measurements from imaging tests or biopsies, such as tumor size, shape, and texture. These features can be used in the logistic function to predict the likelihood of a tumor being diagnosed as benign (0) or malignant (1).

## Implementation

Before we can use Numerical Differentiation or Root Finding methods on our data set, it is first necessary to import the data file from Kaggle.com. Then we will clean up the data set to get rid of the ID number column, and the empty column at the end of the data set.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url = 'https://raw.githubusercontent.com/cjaucian/MAT421/main/data.csv'
data = pd.read_csv(url)

data.drop(['Unnamed: 32', "id"], axis=1, inplace=True)
```

```python
data.head()
```

Before we implement either of our numerical methods, it is necessary to look at the relationships between the predictor variables and the target variable, which is diagnosis.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8-poster')

# Create a list of DataFrame columns
columns = list(data.columns)

# Selecting a subset for analysis
feature_columns = [col for col in data.columns if 'mean' in col]

n_features = len(feature_columns)
n_rows = n_features
fig, ax = plt.subplots(n_rows, 2, figsize=(15, 5 * n_rows))
plt.subplots_adjust(hspace=0.5)

for i, column in enumerate(feature_columns):
    # Boxplot for each feature
    sns.boxplot(x='diagnosis', y=column, data=data, ax=ax [i,0])
    ax[i,0].set_title(f'Boxplot of {column}')

    # Scatterplot for each feature against 'diagnosis'
    sns.scatterplot(x=column, y='area_mean', data=data, hue='diagnosis', palette='viridis', ax=ax[i,1])
    ax[i, 1].set_title(f'Scatterplot of {column} vs. Area Mean')

plt.tight_layout()
plt.show()
```

Based on the data plots, it is clear that numerical differentiation is not going to work with our dataset. The data points do not create a smooth function. This could be circumvented by using interpolation, but the data points are not in ascending order so they fail that method as well.

Looking at the data relationships, we will try a regression model to help make diagnostic predictions.

**Python Method - Logistic Regression**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix
from sklearn.linear_model import LogisticRegression

# Changing the diagnostic column into a numeric value
data_encoded = data.copy()
data_encoded['diagnosis'] = data_encoded['diagnosis'].map({'M' : 1, 'B' : 0})

y = data_encoded['diagnosis']
X = data_encoded.drop('diagnosis', axis=1)

# Create separate data sets for training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize
log_reg = LogisticRegression(max_iter=10000)

# Create a fit on the model with the training data
log_reg.fit(X_train, y_train)
```

```python
# Create predictions on the test set
y_pred = log_reg.predict(X_test)

# Calculate the performance data
accuracy = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

accuracy, recall, precision, f1, conf_matrix
```

```python
print(f'Accuracy: {accuracy:.4f}')
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cbar=False)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

Based on the Confusion Matrix, the logistic regression model paired with machine learning was 97.66% accurate in predicting the diagnosis of the patient. It predicted 106 benign diagnoses, with only 2 misdiagnosed patients. It also predicted 61 malignant diagnoses, with only 2 misdiagnosed patients.
